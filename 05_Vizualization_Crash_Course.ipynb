# Matplotlib Basic

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[0,1,2]
y=[100,200,300]
```

```python
plt.plot(x,y)
```

```python
housing = pd.DataFrame({'rooms':[1,1,2,2,2,3,3,3],
                       'price':[100,120,190,200,230,310,330,305]})

housing
```

```python
plt.scatter(housing['rooms'], housing['price'])
```

```python
plt.plot(x,y, color='#911d67', marker='*', markersize=40, linestyle='--')
plt.title('Title')
plt.xlabel('x Label')
plt.ylabel('y Label')
plt.xlim(0,2)
plt.ylim(0,300)

# to don't see the output (just see the figure) add semicolon at the end
```

# Seaborn Basics

```python
import seaborn as sns
```

```python
df=pd.read_csv('@Source/DATA/heart.csv')
df.head()
```

```python
plt.figure(figsize=(12,8))
sns.distplot(df['age'], bins=20, kde=False, color='red')
```

```python
sns.countplot(data=df, x='cp', hue='sex', palette='plasma')
```

```python
plt.figure(figsize=(12,8))
sns.boxplot(data=df, x='target', y='trestbps', hue='sex')
```

```python
plt.figure(figsize=(12,8))
sns.scatterplot(x='chol', y='trestbps', data=df, hue='sex', palette='Dark2', size='age')
```
